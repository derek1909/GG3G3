# Week 2

In Week 2, your task is to formulate the two models as **Hidden Markov Models (HMM)** with *discrete* states. To recap, <br>
a HMM is described by latent variables that are temporal/sequential and form a Markov chain.<br> 
If we denote these variable at time step $t$ by $x_t$, the Markovian property means that conditioned on $x_t$, all future states<br>
$(x_{t+1}, x_{t+2}, \ldots)$ are independent of states at times before $t$.<br>
Colloquially, conditioned on the present, the future is independent of the past.<br> 

In an HMM the Markovian state variables are not directly observerd. Instead at each time, we observe an observed variable, $n_t$, which<br>
only depends of the Markov state, $x_t$, at the same time step. For us this is given by the Poisson distribution describing the <br>
spike emissions, conditioned on the rates, with the latter being determined by $x_t$ (in other words, the rate at time $t$ is a<br>
deterministic instantaneous functions of $x_t$).<br> Figure 2 shows the graphical model for the HMM.  

Here, **we aim to design Markov chains with discrete states to approximate the behavior of the latent variables of the step and ramp models.**<br>
In the case of the step function, the discretization is actually exact, as the model really just has two levels of rates. <br>
It is for the ramp model (which in its original formulation has continuous states, $x_t$) that **a discretization will be an approximation**. 

The reason for formulation as discrete state HMM's is that in this case we can use the powerful and efficient forward-backward algorithm <br>
to calculate (1) the Bayesian *posterior* estimate (the posterior mean) of the state variables, $x_t$, given the observations, $n_t$, and<br>
(2) calculate the model likelihood function,  $P(n_{1:t}|\Theta, M)$, i.e. the probability of the observed spike train conditioned on the model parameters.




## Task 2.1: 
**Forming an HMM approximation to the ramp model**

$\newcommand{\T}{\mathcal{T}}$

The state variable of the (original) ramp model is continuous and because the update rule $x_{t+1} = x_t + \beta dt + \sigma\sqrt{dt}\epsilon_t$<br>
involves the Gaussian variables $\epsilon_t$, the transition probability $P(x_{t+1}| x_{t})$ is Gaussian (with mean and variance possibly depending on $x_t$, as well as the model parameters). Your first sub-task is to work out this distribution.

First, we will approximate $x_t$ (which lies between 0 and 1) to take values on a regular grid of $K$ points going from 0 to 1 (inclusive of both ends). Let's use $K=50$ or $K=100$.<br>
These $K$ points form the disrete states of our approximate discrete-state HMM. We will denote these states by their index $s$ going from 0 to $K-1$.
The corresponding value of $x_t$ is then given by 

$x_t = \frac{s_t}{K-1} \qquad \text{where} \quad s_t \in \{0, \ldots, K - 1\}$.

By evaluating the Gaussian distribution you derived on this grid, form the transition matrix for the Markov chain, defined as 

$
\T_{s,s'} = P(s_{t+1}= s'| s_{t}= s)
$

Note that the state *transitioned to* (i.e. the one at $t+1$) is the column index of the matrix. Thus the rows of $\T$ have to sum up to 1 (as this is a sum over columns).<br> 
You need to **enforce this constraint by hand** after constructing the (intial) matrix using the evaluated Gaussian. <br>

Also note that according to the original model, once the variable $x$ reaches 1, it stays there. So you will need to use a separate way<br> to construct the last row of $\T$, corresponding to transitions out of state $s = K-1$, corresponding to $x=1$.

```python
from models import *
from models2 import *
from inference import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

## Setup ##
T = 100
time_points = np.arange(T)  # 0,1,2,...

time_ms = time_points * 1e3/T
xr = 0.2
Rh = 50
K = 100 # num of state for ramp HMM

dt = 1/T # unit: s => when t=1s, std of ramp model = sigma
#dt = 1e3/T # ms



#beta = 6/(K-1)/dt
beta = 1
sigma = 1e-5#1/(K-1)/dt**0.5 
# st = np.arange(K) # states
# xt = st/(K-1)
st = np.arange(K) # 2K states
xt = st/(K-1)

plt.figure(figsize=(5, 3))


## transition matrix
# calculate difference between states, forming a K x K matrix.
s_grid = st - st.reshape(-1,1) 
mu = beta * dt * (K-1)
std = sigma * np.sqrt(dt) * (K-1)
# print(s_grid)
# print(mu)
# print(std)

# Transition matrix, last row are calculated differently

# # Suppose trans_matrix is in log space
# log_trans_matrix = stats.norm.logpdf(s_grid, mu, std)
# # First column
# log_trans_matrix[:,0] = stats.norm.logcdf(s_grid[:,0], mu, std)
# # Last column
# log_trans_matrix[:,-1] = 1-stats.norm.logcdf(s_grid[:,-1], mu, std)

# Suppose trans_matrix is in log space
log_trans_matrix = np.log(stats.norm.cdf(s_grid+0.5, mu, std) - stats.norm.cdf(s_grid-0.5, mu, std))

# First column
log_trans_matrix[:,0] = stats.norm.logcdf(s_grid[:,0]+0.5, mu, std)
# Last column
log_trans_matrix[:,-1] = stats.norm.logsf(s_grid[:,-1]-0.5, mu, std)

# Last row
log_trans_matrix[-1] = -np.inf
log_trans_matrix[-1,-1] = 0
#print("sum of row (without normalize) = ", np.sum(np.exp(log_trans_matrix),axis=1))


#Compute row sums in log space using logsumexp
log_row_sums = np.zeros(log_trans_matrix.shape[0]) 
for i in range(log_trans_matrix.shape[0]):
    log_row_sums[i] = logsumexp(log_trans_matrix[i, :])
#Subtract row sums from each element in log space to normalize
normalized_log_trans_matrix = log_trans_matrix - log_row_sums[:, np.newaxis]

#print(normalized_log_trans_matrix)
trans_matrix = np.exp(normalized_log_trans_matrix)
#print("sum of row = ", np.sum(trans_matrix,axis=1))

plt.title(f"Visualisation of the state transition matrix, K={K}")
im = plt.imshow(trans_matrix,extent=[K,-K, K,-K])#,vmin=0, vmax=1)
plt.colorbar(im)


# Change last row to correct value
# trans_matrix[-1] = 0
# trans_matrix[-1,-1] = 1
# row_sums = trans_matrix.sum(axis=1)
# trans_matrix = trans_matrix / row_sums[:, np.newaxis]
#print(np.sum(trans_matrix[2]))

# print(trans_matrix)
#plt.plot(trans_matrix[4])
```

Next, you will need to form the initial state distribution, $\pi$, as an array of $K$ values (summing to 1!) giving the probabilities of different possibilities of $s_0$. This should approximate the equation $x[0] = x_0 + \sigma\sqrt{dt} \epsilon_0$.

```python
mu = xr * (K-1)
std = sigma * np.sqrt(dt) * (K-1)
#Suppose trans_matrix is in log space
# log_pi = stats.norm.logpdf(st, mu, std)
# log_pi[0] = stats.norm.logcdf(st[0], mu, std)
# log_pi[-1] = stats.norm.logsf(st[-1], mu, std) # sf = 1-cdf

log_pi = np.log(stats.norm.cdf(st+0.5, mu, std) - stats.norm.cdf(st-0.5, mu, std))
log_pi[0] = stats.norm.logcdf(st[0]+0.5, mu, std)
log_pi[-1] = stats.norm.logsf(st[-1]-0.5, mu, std) # sf = 1-cdf

#Compute row sums in log space using logsumexp
log_pi_sums = logsumexp(log_pi)
#Subtract row sums from each element in log space to normalize
normalized_log_pi = log_pi - log_pi_sums
# recover the pi
pi = np.exp(normalized_log_pi)
```

Once the transition matrix $\T$ and initial state distribution $\pi$ are formed, we can simulate the chain. In order to <br> do this, you will first draw $s_0$ from the initial state distribution, then successively sample from the <br>
appropriate distribution according to the transition matrix $\T$ (and depending on the current state $s_t$). To sample <br>
the discrete (integer) $s$ from a distribution over its $K$ possibilities, you can use `np.random.choice`.

```python
## Simulate ramp model using HMM

# 1xT matrix to record states
states = np.zeros(T, dtype=int)

# Draw the initial state from the initial distribution
# K states, pi initial distribution. => return a scaler with initial values (discrete)
states[0] = np.random.choice(st, size=None, replace=True, p=pi)

# Simulate the chain
for t in range(1, T):
    # The transition probabilities depend on the current state
    current_state = states[t-1]
    states[t] = np.random.choice(st, size=None, replace=True, p=trans_matrix[current_state])


Xt = states / (K-1)
plt.plot(time_ms, Xt * Rh)

plt.ylabel("spike rate (sp/s)")
plt.xlabel("time from motion onset (ms)")
plt.ylim(0,55)
plt.xlim(0,1000)
plt.show()
```

```python
from models import *
from models2 import *
from inference import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import time


## Setup ##
T = 100
time_points = np.arange(T) # 0,1,2,...

dt = 1/T
time_ms = time_points * dt * 1e3
r = 25
m = 50
p = r/(m+r)
Rh = 50
xs = 0.2
Rh = 50

plt.figure(figsize=(5, 3))

#### initial prob distribution is [1,0], but for NB distribution,#failure=0 is possible <=> jump at t=0 is possible 
#### => prob_h[0] should be [1,0] * trans


## 1. Inhomogeneous markov chain ##
PMF_jump = stats.nbinom.pmf(time_points, r, p)
CMF_jump = stats.nbinom.cdf(time_points, r, p)

Xt = CMF_jump + (1-CMF_jump) * xs 
plt.plot(time_ms, Xt, label="CMF of jump time",linewidth=0.7)

trans_matrices = np.empty((T-1,2,2)) # inhomogeneous transition matrix
for t in range(0,T-1):
    if CMF_jump[t]==1:
        trans_matrices[t] = np.array([[ 0, 1], [0,1]])
    else:
        trans_matrices[t] = np.array([[ (1-CMF_jump[t+1]) / (1-CMF_jump[t]), PMF_jump[t+1]/(1-CMF_jump[t]) ], [0,1]])
    
# Prob of observed state at time t = [prob of state=0, prob of state=1]
prob_h = np.zeros((T,2))
pi = [1,0] # initial state

prob_h[0] = np.matmul(pi,np.array([[1-PMF_jump[0],PMF_jump[0]],[0,1]])) # P(t=0)
for t in range(0,T-1):
    prob_h[t+1] = np.matmul(prob_h[t],trans_matrices[t])
Xt = prob_h[:,1] + prob_h[:,0] * xs 
plt.plot(time_ms, Xt, label="prob of xt=1, inhomogeneous MC ",linewidth=2, linestyle='-.')

#print(trans_matrices)



## 2. 2-states homogeneous markov chain ##

trans_matrix = np.array([[1- 1/m, 1/m],[0,1]])
pi = [1,0] # initial state

prob_n = np.zeros((T,2))
prob_n[0]=np.matmul(pi,trans_matrix)
for ii in range(1,T):
    prob_n[ii] =np.matmul(prob_n[ii-1],trans_matrix)
    
Xt = prob_n[:,1] + prob_n[:,0] * xs 
# plt.plot(time_ms, Xt, label="prob of xt=1, 2-state MC",linewidth=2,linestyle='-')
#print(trans_matrix)



## 3. r+1 states MC (1) ##

# Xt can be: [0,1,2,...,r]
# With probability p, Xt increase by one
# When Xt at its r+1 th state, nt=1, otherwise nt=0

# Find transition matrix for r and p
trans_matrix = np.zeros((r+1,r+1))
trans_matrix[0, 0:2] = [1-p, p] # Set the first row
for i in range(1, r): # Create the remaining rows
    trans_matrix[i, i:i+2] = [1-p, p]
trans_matrix[r , r ] = 1 # Set the last row
# Find initial distribution of latent state 
pi = np.zeros(r+1)
pi[0] = 1


# simulate latent variable 
prob_h = np.zeros((T+r,r+1)) # prob of hidden states
prob_h[0] = np.matmul(pi, trans_matrix)
for ii in range(1,T+r):
    prob_h[ii] =np.matmul(prob_h[ii-1], trans_matrix)
    #print(f"prob_h[{ii}]: ", prob_h[ii])


# Calculate Xt=(r+1)th state from latent variable
# index = 0 means after 1 steps; index = n means after n+1 steps;
# high at n+1 steps <=> jump time = n+1-r <=> high at t_index=n+1-r 
# i.e. should shift r-1
Xt = np.roll(prob_h[:,r], shift=-r+1)[:T] + (1-np.roll(prob_h[:,r], shift=-r+1)[:T]) * xs # shift r because r success experiment should not be included
# plt.plot(time_ms, Xt, label="prob of xt=1, (r+1)-state MC (1)",linewidth=2,linestyle=':')



## 4. r+1 states MC (2) ##

# Xt can be: [0,1,2,...,r]
# With probability p, Xt increase by one
# When Xt at its r+1 th state, nt=1, otherwise nt=0

# Find transition matrix for r and p
trans_matrix = np.zeros((r+1, r+1))
for i in range(r+1): # axis=0
    for j in range(i, r+1): # axis=1
        trans_matrix[i, j] = (p**(j-i)) * (1-p) if j < r else p**(r-i)
# Find initial distribution of latent state 
pi = np.zeros(r+1)
pi[0] = 1


# simulate latent variable 
prob_h = np.zeros((T,r+1)) # prob of hidden states
prob_h[0] = np.matmul(pi, trans_matrix)
print(prob_h[0] )
for ii in range(1,T):
    prob_h[ii] =np.matmul(prob_h[ii-1], trans_matrix)
    #print(f"prob_h[{ii}]: ", prob_h[ii])


# Calculate Xt=(r+1)th state from latent variable
Xt = prob_h[:,r] * (1-xs) + xs 

# plt.plot(time_ms, Xt, label="prob of xt=1, (r+1)-state MC (2)",linewidth=2,linestyle=':')

# Plot setup
plt.legend()
plt.title("Step model simulation using different markov chain model")

plt.ylabel("x_t")
plt.xlabel("time from motion onset (ms)")
plt.ylim(0,1.05)
plt.xlim(0,1000)
#plt.title(f"PSTH for RAMP model, sigma={s}, beta={b}, bin size = 20")
#filename = f"Week2/2.2/PSTH_RAMP_s{s}_beta{b}.png"
#plt.savefig(filename,bbox_inches='tight')
#plt.show()
```

- The `hmm_expected_states` function has an optional boolean input `filter` which is false by default. When true, the function calculates $P(s_t | n_{1:t})$ instead of $P(s_t | n_{1:T})$: i.e. the posterior conditioned only on observations up to and including the "current" time-step $t$. This corresponds to the so-called filtering problem (as in the Kalman filter; the default case, `filter = False`, corresponds to "smoothing". Filtering is appropriate for applications where inference has to be performed online, in which case, to infer $x_t$ we do not have the luxury of having access to future observations -- without a time-machine, that is! In our case (as engineers studying the computational mechanisms in area LIP), we do of course have access to the entire spike-train; hence the default option. **Nevertheless, carry out a theoretical study of the differences in inference accuracy, for both models, using smoothing vs. filtering.
```python
# Ramp model ll maps

from models import *
from models2 import *
from inference import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.ticker as ticker

import time

start_time = time.time()

## Setup ##
M=20
values_logs = np.linspace(np.log(0.04), np.log(4), M) # -3.22 - 1.386
values_b = np.linspace(0, 4, M)
values_x0 = np.linspace(0, 1, M)



T=100
time_points = np.linspace(1,T,T) # 0,1,2,...
dt = 1/T
time_ms = time_points * dt * 1e3

Rh = 50
xr = 0.2
bin_size = 20
bin_size_2 = 50
bin_edges = np.arange(0, 1e3+bin_size, bin_size)

# plt.figure(figsize=(5, 3))

# Dataset
actual_b = 1.3
actual_logs = np.log(1)
actual_x0 = 0.9
N = 16
K = 20
st = np.arange(K) # states
xt = st/(K-1)

ramp = RampModel(beta=actual_b, sigma=np.exp(actual_logs), x0=actual_x0, Rh=50)
[spikes, _, _, _, _] = ramp.simulate_HMM(Ntrials=N, T=T, K=K, get_rate=True)

counts_matrix = generate_psth(spike_trains=spikes, bin_size=bin_size, bin_size_2=bin_size_2, plot=True, return_counts=True);


log_prior = np.log(1/M**3)
model_ll = np.zeros((M,M,M))

for b_idx in range(M):
    for s_idx in range(M):
        for x_idx in range(M):

            b = values_b[b_idx]
            logs = values_logs[s_idx]
            x0 = values_x0[x_idx]

            ramp = RampModel(beta=b, sigma=np.exp(logs), x0=x0, Rh=50)
            [_, _, _, normalized_log_trans_matrix, normalized_log_pi0] = ramp.simulate_HMM(Ntrials=0, 
                                                                                           T=T, 
                                                                                           K=K, 
                                                                                           get_rate=True)
            normalized_log_trans_matrix = normalized_log_trans_matrix[np.newaxis, :, :] # (1,K,K) for homogeneous MC

            lls = poisson_logpdf(counts=counts_matrix, lambdas= xt*Rh*dt, mask=None) # N x T x K 

            # Model log-likelihood for N trials:
            for n in range(N):
                model_ll[b_idx, s_idx, x_idx] += hmm_normalizer(log_pi0 = normalized_log_pi0, 
                                                                log_Ps = normalized_log_trans_matrix, 
                                                                ll = lls[n])


unnormalised_log_poste = model_ll + log_prior
log_poste = unnormalised_log_poste - logsumexp_scipy(unnormalised_log_poste)

poste = np.exp(log_poste) + 1e-16
poste_b = np.sum(poste,axis=(1,2)) # b is axis0, so sum over axis 1,2
poste_logs = np.sum(poste,axis=(0,2))
poste_x0 = np.sum(poste,axis=(0,1))

expected_b = np.sum(values_b * poste_b)
var_b = np.sum(poste_b * (values_b - expected_b)**2)
std_b = np.sqrt(var_b)

expected_logs = np.sum(values_logs * poste_logs)
var_logs = np.sum(poste_logs * (values_logs - expected_logs)**2)
std_logs = np.sqrt(var_logs)

expected_x0 = np.sum(values_x0 * poste_x0)
var_x0 = np.sum(poste_x0 * (values_x0 - expected_x0)**2)
std_x0 = np.sqrt(var_x0)
        

end_time = time.time()
elapsed_time = end_time - start_time

print("Takes", elapsed_time, "s")     
```
